```python
class Person:
    def __init__(self, name, age, ht):
        self.name = name
        self.age = age
        self.height = ht

    def show_info(self):
        return f"name -> {self.name}, age -> {self.age}, height -> {self.height}"
```

```python
p1 = Person('aarya', 17, 5.6)
p2 = Person('akshata', 17, 5.7)
```

```python
p1.show_info()
```

```python
class Employee:
    company = "ABC Company"

    def __init__(self, name, monthly_salary, job_designation):
        self.name = name
        self.__monthly_salary = monthly_salary
        self.job_designation = job_designation

    def annual_salary(self):
        print("Annual Salary =", self.__monthly_salary * 12)

    def info(self):
        return f"Name={self.name}, Salary={self.__monthly_salary}, Designation={self.job_designation}, Company={Employee.company}"


e1 = Employee("Aarya", 30000, "Engineer")
e2 = Employee("Akshata", 40000, "Manager")
e3 = Employee("Shreya", 50000, "Analyst")

print(e1.info())
e1.annual_salary()

print(e2.info())
e2.annual_salary()

print(e3.info())
e3.annual_salary()
```
