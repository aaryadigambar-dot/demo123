```python
class BankAccount:
    def __init__(self,name,acc_no,balance):
        self.name = name
        self.acc_no = acc_no
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balnace")
        else:
            self.balance -= amount

    def get_balance(self):
        return self.balnance

    def __str__(self):
        return f"Name: {self.name}, Account No: {self.acc_no}, Balance: {self.balance}"

a1 = BankAccount("Aarya", 101, 5000)
a1.deposit(1000)
a1.withdraw(2000)

a2 = BankAccount("Rahul", 102, 2000)
a2.deposit(500)
a2.withdraw(4000)

print(a1)
print(a2)
```

```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def calculate_percentage(self):
        return sum(self.marks) / len(self.marks)

    def get_grade(self):
        p = self.calculate_percentage()

        if p >= 80:
            return "A"
        elif p >= 60:
            return "B"
        elif p >= 40:
            return "C"
        else:
            return "Fail"

    def display(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)
        print("Percentage:", self.calculate_percentage())
        print("Grade:", self.get_grade())
        print()


s1 = Student("Aarya", 1, [80, 85, 90, 75, 70])
s2 = Student("Rahul", 2, [60, 65, 70, 55, 75])
s3 = Student("Priya", 3, [30, 35, 40, 45, 50])

students = [s1, s2, s3]

for i in students:
    i.display()
```

```python
class Employee:
    def __init__(self, name, emp_id, salary):
        self.__name = name
        self.__emp_id = emp_id
        self.__salary = salary

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        if salary < 10000:
            raise ValueError("Salary must be above 10000")
        self.__salary = salary

    def give_raise(self, percent):
        self.__salary += self.__salary * percent / 100


e = Employee("Aarya", 101, 20000)

print("Salary:", e.get_salary())

e.give_raise(10)
print("New Salary:", e.get_salary())

try:
    e.set_salary(5000)
except ValueError as e:
    print(e)

try:
    print(e.__salary)
except:
    print("Cannot access private attribute directly")
```

```python
class SecureVault:
    def __init__(self, pin):
        self.__pin = pin
        self.__data = {}

    def add_entry(self, pin, key, value):
        if pin == self.__pin:
            self.__data[key] = value
        else:
            print("Access Denied")

    def get_entry(self, pin, key):
        if pin == self.__pin:
            return self.__data.get(key)
        else:
            print("Access Denied")


v = SecureVault(1234)

v.add_entry(1234, "Name", "Aarya")
print(v.get_entry(1234, "Name"))

v.get_entry(1111, "Name")
```

```python
class Shape:
    def area(self):
        return 0

    def describe(self):
        print("I am a shape")


class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return 3.14 * self.r * self.r

    def describe(self):
        print("I am a Circle")


class Rectangle(Shape):
    def __init__(self, l, w):
        self.l = l
        self.w = w

    def area(self):
        return self.l * self.w

    def describe(self):
        print("I am a Rectangle")


class Triangle(Shape):
    def __init__(self, b, h):
        self.b = b
        self.h = h

    def area(self):
        return 0.5 * self.b * self.h

    def describe(self):
        print("I am a Triangle")


c = Circle(5)
r = Rectangle(4, 6)
t = Triangle(4, 8)

c.describe()
print("Area =", c.area())

r.describe()
print("Area =", r.area())

t.describe()
print("Area =", t.area())
```

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started")


class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def start_engine(self):
        print(self.model, "Engine started")


class ElectricCar(Vehicle):
    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)
        self.battery_capacity = battery_capacity

    def start_engine(self):
        print("Silent start - battery at", self.battery_capacity, "kWh")


c = Car("Toyota", "Fortuner", 2023, 4)
e = ElectricCar("Tesla", "Model 3", 2024, 75)

c.start_engine()
e.start_engine()
```

```python
class Animal:
    def sound(self):
        print("Some sound")


class Dog(Animal):
    def sound(self):
        print("Bark")


class Cat(Animal):
    def sound(self):
        print("Meow")


class Cow(Animal):
    def sound(self):
        print("Moo")


def make_all_sounds(animals):
    for a in animals:
        a.sound()


animals = [Dog(), Cat(), Cow()]
make_all_sounds(animals)
```

```python
class Calculator:
    def add(self, a, b, c=0):
        return a + b + c

    def multiply(self, a, b, c=1):
        return a * b * c


c = Calculator()

print(c.add(2, 3))
print(c.add(2, 3, 4))

print(c.multiply(2, 3))
print(c.multiply(2, 3, 4))


class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def show(self):
        print(self.x, self.y)


v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2
v3.show()
```
