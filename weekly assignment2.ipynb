```python
def analyze_sentence(sentence):
    words = sentence.lower().split()

    print("Word Count =", len(words))
    print("Character Count =", len(sentence.replace(" ", "")))
    print("Unique Words =", len(set(words)))

    most_word = max(words, key=words.count)
    print("Most Frequent Word =", most_word)

sentence = "The quick brown fox jumps over the lazy dog the fox"
analyze_sentence(sentence)
```

```python
text = "A man a plan a canal Panama"

# a) Reverse the string
reverse = ""

for i in text:
    reverse = i + reverse

print(reverse)
```

```python
# b) Check palindrome
text2 = text.lower().replace(" ", "")

reverse = ""

for i in text2:
    reverse = i + reverse

if text2 == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# c) Replace vowels with *
new_text = text

new_text = new_text.replace("a", "*")
new_text = new_text.replace("A", "*")
new_text = new_text.replace("e", "*")
new_text = new_text.replace("i", "*")
new_text = new_text.replace("o", "*")
new_text = new_text.replace("u", "*")

print(new_text)
```

```python
def format_report(data):
    print(f"{'Name':<15}{'Score':>6}{'Grade':^8}")

    for name, score, grade in data:
        print(f"{name:<15}{score:>6}{grade:^8}")

data = [
    ("Alice", 92, "A"),
    ("Bob", 78, "B+"),
    ("Charlie", 55, "C"),
    ("Diana", 100, "A+")
]

format_report(data)
```

```python
def list_ops(nums):
    # a) Remove duplicates
    unique = []
    for i in nums:
        if i not in unique:
            unique.append(i)
    print("Without duplicates:", unique)

    # b) Even index elements squared
    squared = [nums[i]**2 for i in range(0, len(nums), 2)]
    print("Even index squared:", squared)

    # c) Flatten nested list
    nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
    flat = [j for i in nested for j in i]
    print("Flattened list:", flat)

nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]

list_ops(nums)
```

```python
employee = ("E042", "Riya Sharma", "Data Science", 72000, "Pune")

# a) Unpacking
id, name, dept, salary, city = employee

print(id)
print(name)
print(dept)
print(salary)
print(city)

# b) * unpacking
first, *middle, last = employee

print(first)
print(middle)
print(last)

# c) Update salary
def update_salary(emp, new_salary):
    return emp[:3] + (new_salary,) + emp[4:]

print(update_salary(employee, 80000))
```

```python
batch_a = {"Aarav", "Aarya", "Karan", "Akshata", "Sahil"}
batch_b = {"Karan", "Meera", "Tanvi", "Rohan"}
batch_c = {"Priya", "Aarav"}

# a) Common students
print("a)", batch_a & batch_b)

# b) In batch_a but not in batch_b
print("b)", batch_a - batch_b)

# c) All unique students
print("c)", batch_a | batch_b)

# d) Check subset
print("d)", batch_c.issubset(batch_a))
```

```python
def inventory_manager(records):
    result = {}

    for product, category, quantity, price in records:
        if category not in result:
            result[category] = [0, 0, []]

        result[category][0] += quantity
        result[category][1] += quantity * price
        result[category][2].append(product)

    return result


records = [
    ("Laptop", "Electronics", 10, 55000),
    ("Mouse", "Electronics", 50, 499),
    ("Desk", "Furniture", 8, 12000),
    ("Chair", "Furniture", 15, 4500),
    ("Pen", "Stationery", 200, 20)
]

print(inventory_manager(records))
```

```python
# a) power_all
def power_all(*args, exp=2):
    result = []
    for i in args:
        result.append(i ** exp)
    return result

print(power_all(2, 3, 4))


# b) profile
def profile(**kwargs):
    for k, v in kwargs.items():
        print(k, ":", v)

profile(name="Arjun", age=20)


# c) sort by cgpa
students = [
    {"name": "Arjun", "cgpa": 8.4},
    {"name": "Sanya", "cgpa": 9.1},
    {"name": "Dev", "cgpa": 7.8}
]

students.sort(key=lambda x: x["cgpa"], reverse=True)

print(students)
```

```python
import math

# a)
def make_multiplier(n):
    def multiply(x):
        return x * n
    return multiply

triple = make_multiplier(3)
print(triple(4))


# b)
def timer(func):
    def wrapper(n):
        print("Function Running...")
        return func(n)
    return wrapper

@timer
def slow_sum(n):
    total = 0
    for i in range(1, n + 1):
        total += i * i
    return total

print(slow_sum(5))


# c)
def validate_positive(func):
    def wrapper(x):
        if x > 0:
            return func(x)
        else:
            print("Invalid Number")
    return wrapper

@validate_positive
def safe_sqrt(x):
    return math.sqrt(x)

print(safe_sqrt(25))
```
