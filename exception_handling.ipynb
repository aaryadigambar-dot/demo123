```python
# try and except
```

```python
try:
    f = open('all.txt')
except:
    print("file does not exist")
```

```python
try:
    f = open('a1.txt')
    print(var1)
except:
    print("file does not exist")
```

```python
# we need to specific about what kind of error we want to catch
```

```python
try:
    f = open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("file does not exist")
```

```python
try:
    f = open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("file does not exist")
except Exception:
    print("something went wrong")
```

```python
try:
    f = open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
```

```python
# else -> idf try block dosen't raise any exception this will run
```

```python
try:
    f = open('a1.txt')
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
```

```python
# finally
```

```python
try:
    f = open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("run complete")
```

```python
# Manaually 
```

```python
try:
    f = open('a1.txt')
    if f.name == 'a1.txt':
       raise Exception
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("run complete")
```

```python
# take 2 numbers as user input
# write a program for division of these 2 numbers 
# handle zerodivisionerror and valueerror
```

```python
try:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))

    result = a / b
    print("Division =", result)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Please enter valid numbers")
```
